Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python
file_path = '/content/drive/MyDrive/rumus tugas excel/StudentsPerformance.csv'

# Baca dataset ke dalam DataFrame
df = pd.read_csv(file_path)

# Pilih kolom bertipe data kategorikal (object)
categorical_columns = df.select_dtypes(include=['object']).columns

# Membuat pie chart untuk setiap kolom kategorikal
for col in categorical_columns:
    # Hitung distribusi frekuensi
    value_counts = df[col].value_counts()

    # Membuat pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

    # Tambahkan judul sesuai kolom
    plt.title(f'Distribution of {col}')

    # Tampilkan chart
    plt.axis('equal')  # Membuat pie chart menjadi lingkaran sempurna
    plt.show()
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python


# Baca dataset ke dalam DataFrame
df = pd.read_csv(file_path)

# Menambah kolom baru untuk 'status' lulus atau gagal berdasarkan nilai 75
df['math_status'] = df['math score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['reading_status'] = df['reading score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['writing_status'] = df['writing score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')

# Membuat plot ukuran
plt.figure(figsize=(15, 5))

# Plot untuk Math
plt.subplot(1, 3, 1)
sns.countplot(x='math_status', data=df, palette='coolwarm')
plt.title('Pass vs Fail in Math')
plt.xlabel('Math Status')
plt.ylabel('Number of Students')

# Plot untuk Reading
plt.subplot(1, 3, 2)
sns.countplot(x='reading_status', data=df, palette='coolwarm')
plt.title('Pass vs Fail in Reading')
plt.xlabel('Reading Status')
plt.ylabel('Number of Students')

# Plot untuk Writing
plt.subplot(1, 3, 3)
sns.countplot(x='writing_status', data=df, palette='coolwarm')
plt.title('Pass vs Fail in Writing')
plt.xlabel('Writing Status')
plt.ylabel('Number of Students')

# Tampilkan plot
plt.tight_layout()
plt.show()
````

Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python

# Baca dataset ke dalam DataFrame
df = pd.read_csv(file_path)

# Set ukuran plot
plt.figure(figsize=(20, 15))

# 1. Pengaruh gender terhadap nilai math, reading, dan writing
plt.subplot(3, 1, 1)
sns.boxplot(x='gender', y='math score', data=df, palette='coolwarm')
plt.title('Gender vs Math Score')

plt.subplot(3, 1, 2)
sns.boxplot(x='gender', y='reading score', data=df, palette='coolwarm')
plt.title('Gender vs Reading Score')

plt.subplot(3, 1, 3)
sns.boxplot(x='gender', y='writing score', data=df, palette='coolwarm')
plt.title('Gender vs Writing Score')

plt.tight_layout()
plt.show()

# Visualisasi untuk Race/Ethnicity, Parental Level of Education, Lunch, dan Test Preparation Course
factors = ['race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

for factor in factors:
    plt.figure(figsize=(20, 15))

    # 2. Pengaruh terhadap math score
    plt.subplot(3, 1, 1)
    sns.boxplot(x=factor, y='math score', data=df, palette='coolwarm')
    plt.title(f'{factor.capitalize()} vs Math Score')
    plt.xticks(rotation=45)

    # 3. Pengaruh terhadap reading score
    plt.subplot(3, 1, 2)
    sns.boxplot(x=factor, y='reading score', data=df, palette='coolwarm')
    plt.title(f'{factor.capitalize()} vs Reading Score')
    plt.xticks(rotation=45)

    # 4. Pengaruh terhadap writing score
    plt.subplot(3, 1, 3)
    sns.boxplot(x=factor, y='writing score', data=df, palette='coolwarm')
    plt.title(f'{factor.capitalize()} vs Writing Score')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
````

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python

# Gunakan path yang benar untuk file Anda
file_path = '/content/drive/MyDrive/rumus tugas excel/StudentsPerformance.csv'

# Baca dataset ke dalam DataFrame
df = pd.read_csv(file_path)

# Membuat kolom baru untuk rata-rata nilai dari math, reading, dan writing
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Set ukuran default plot
plt.figure(figsize=(20, 15))

# Visualisasi korelasi antara rata-rata nilai dan kategori: gender, race/ethnicity, parental level of education, lunch, test preparation course
factors = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

for factor in factors:
    plt.figure(figsize=(10, 6))

    # Membuat boxplot untuk rata-rata nilai berdasarkan kategori
    sns.boxplot(x=factor, y='average_score', data=df, palette='coolwarm')

    # Set judul dan label
    plt.title(f'Average Score vs {factor.capitalize()}')
    plt.xlabel(factor.capitalize())
    plt.ylabel('Average Score')

    # Rotate label x jika diperlukan
    plt.xticks(rotation=45)

    # Tampilkan plot
    plt.tight_layout()
    plt.show()

````
