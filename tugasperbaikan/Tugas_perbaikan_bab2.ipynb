```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.svm import SVC, SVR
from sklearn.preprocessing import StandardScaler
```

```python
# Load dataset (Iris for classification, California Housing for regression)
iris = datasets.load_iris()
california = datasets.fetch_california_housing()

# Split data for classification
X_classification = iris.data
y_classification = iris.target
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_classification, y_classification, test_size=0.3, random_state=42)

# Split data for regression (California housing dataset)
X_regression = california.data
y_regression = california.target
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_regression, y_regression, test_size=0.3, random_state=42)
```

```python
# Standardize the features (important for SVM, k-NN)
scaler = StandardScaler()
X_train_class = scaler.fit_transform(X_train_class)
X_test_class = scaler.transform(X_test_class)

X_train_reg = scaler.fit_transform(X_train_reg)
X_test_reg = scaler.transform(X_test_reg)
```

```python
# 1. k-Nearest Neighbors (k-NN) for Classification
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_class, y_train_class)
y_pred_class_knn = knn.predict(X_test_class)
accuracy_knn = accuracy_score(y_test_class, y_pred_class_knn)
print(f'k-NN Accuracy: {accuracy_knn:.3f}')
```

```python
# 2. Logistic Regression (Linear Model) for Classification
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train_class, y_train_class)
y_pred_class_log_reg = log_reg.predict(X_test_class)
accuracy_log_reg = accuracy_score(y_test_class, y_pred_class_log_reg)
print(f'Logistic Regression Accuracy: {accuracy_log_reg:.3f}')
```

```python
# 3. Decision Tree Classifier for Classification
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train_class, y_train_class)
y_pred_class_dt = dt_classifier.predict(X_test_class)
accuracy_dt = accuracy_score(y_test_class, y_pred_class_dt)
print(f'Decision Tree Accuracy: {accuracy_dt:.3f}')
```

```python
# 4. Random Forest Classifier for Classification
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_class, y_train_class)
y_pred_class_rf = rf_classifier.predict(X_test_class)
accuracy_rf = accuracy_score(y_test_class, y_pred_class_rf)
print(f'Random Forest Accuracy: {accuracy_rf:.3f}')
```

```python
# 5. Naive Bayes Classifier for Classification
nb_classifier = GaussianNB()
nb_classifier.fit(X_train_class, y_train_class)
y_pred_class_nb = nb_classifier.predict(X_test_class)
accuracy_nb = accuracy_score(y_test_class, y_pred_class_nb)
print(f'Naive Bayes Accuracy: {accuracy_nb:.3f}')
```

```python
# 6. Support Vector Machine (SVM) for Classification
svm_classifier = SVC(kernel='linear', random_state=42)
svm_classifier.fit(X_train_class, y_train_class)
y_pred_class_svm = svm_classifier.predict(X_test_class)
accuracy_svm = accuracy_score(y_test_class, y_pred_class_svm)
print(f'SVM Accuracy: {accuracy_svm:.3f}')
```

```python
# 7. Linear Regression for Regression (California dataset)
lr_regressor = LinearRegression()
lr_regressor.fit(X_train_reg, y_train_reg)
y_pred_reg_lr = lr_regressor.predict(X_test_reg)
mse_lr = mean_squared_error(y_test_reg, y_pred_reg_lr)
print(f'Linear Regression MSE: {mse_lr:.3f}')
```

```python
# 8. Decision Tree Regressor for Regression
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train_reg, y_train_reg)
y_pred_reg_dt = dt_regressor.predict(X_test_reg)
mse_dt = mean_squared_error(y_test_reg, y_pred_reg_dt)
print(f'Decision Tree Regressor MSE: {mse_dt:.3f}')
```

```python
# 9. Random Forest Regressor for Regression
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train_reg, y_train_reg)
y_pred_reg_rf = rf_regressor.predict(X_test_reg)
mse_rf = mean_squared_error(y_test_reg, y_pred_reg_rf)
print(f'Random Forest Regressor MSE: {mse_rf:.3f}')
```

```python
# 10. Support Vector Machine (SVM) for Regression
svm_regressor = SVR(kernel='linear')
svm_regressor.fit(X_train_reg, y_train_reg)
y_pred_reg_svm = svm_regressor.predict(X_test_reg)
mse_svm = mean_squared_error(y_test_reg, y_pred_reg_svm)
print(f'SVM Regressor MSE: {mse_svm:.3f}')
```

```python
# Plot confusion matrix for classification (k-NN)
plt.figure(figsize=(6, 6))
conf_matrix_knn = confusion_matrix(y_test_class, y_pred_class_knn)
plt.imshow(conf_matrix_knn, cmap='Blues', interpolation='nearest')
plt.title("Confusion Matrix - k-NN")
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
# Plot predicted vs actual for regression (Linear Regression)
plt.figure(figsize=(6, 6))
plt.scatter(y_test_reg, y_pred_reg_lr)
plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], color='red', lw=2)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True vs Predicted - Linear Regression')
plt.show()
```
